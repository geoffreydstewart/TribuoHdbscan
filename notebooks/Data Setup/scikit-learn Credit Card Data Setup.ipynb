```python
# on the other hand, the missing values in MINIMUM_PAYMENTS can likely be safely replaced with 0.

df.fillna(0, inplace=True)
```

```python
# Instantiate a dataframe, so we can keep the original one as is for future reference, before scaling the data

df_X = pd.DataFrame(df)
```

```python
df_X.drop(['CUST_ID'], axis=1, inplace=True)
```

```python
# scale these numeric values using a MinMax scaler. The min/max values are well defined in this case.
x_scaler = MinMaxScaler()
df_X[['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE']] = x_scaler.fit_transform(df_X[['BALANCE', 'BALANCE_FREQUENCY', 'PURCHASES', 'ONEOFF_PURCHASES', 'INSTALLMENTS_PURCHASES', 'CASH_ADVANCE', 'PURCHASES_FREQUENCY', 'ONEOFF_PURCHASES_FREQUENCY', 'PURCHASES_INSTALLMENTS_FREQUENCY', 'CASH_ADVANCE_FREQUENCY', 'CASH_ADVANCE_TRX', 'PURCHASES_TRX', 'CREDIT_LIMIT', 'PAYMENTS', 'MINIMUM_PAYMENTS', 'PRC_FULL_PAYMENT', 'TENURE']])

# Required for the second round of feature reduction
# df_X[['odometer_value', 'year_produced']] = x_scaler.fit_transform(df_X[['odometer_value', 'year_produced']])
df_X.head(5)
```
